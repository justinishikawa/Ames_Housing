```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
from datetime import timedelta
```

```python
df=pd.read_pickle("Day_Age_complaint")
```

```python
plt.hist(df.Age_Days.astype('timedelta64[D]')
        )
np.max(df.Age_Days.astype('timedelta64[D]'))
```

We see from the chart above that the majority of complaints are about items that are less that 500 days from the date of manufacturing. Also notice the skewness of the data with the right tail gone as far as 10,139 days after being manufactured
